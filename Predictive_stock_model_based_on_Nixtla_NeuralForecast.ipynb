```python
DATA_DIR = r"C:\Users\pawel\Documents\Python nauka\Projekt\Notowania\stocks_1d"

def load_stock_data_for_symbols(symbols, data_dir=DATA_DIR):

    all_data = {}

    for symbol in symbols:
        file_path = os.path.join(data_dir, f"{symbol}.txt")

        if not os.path.exists(file_path):
            print(f"[BD] Brak pliku dla symbolu: {symbol}")
            continue

        try:
            column_names = ["Date", "Open", "High", "Low", "Close", "Volume"]
            df = pd.read_csv(file_path, sep=",", header=None, names=column_names)

            df.dropna(subset=["Date"], inplace=True)
            df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
            df.set_index("Date", inplace=True)
            df = df[df.index >= "2010-01-01"]

            if df.empty:
                print(f"[BD] Brak danych po filtracji dla: {symbol}")
                continue

            print(f"[OK] Wczytano dane dla: {symbol}")
            all_data[symbol] = df

        except Exception as e:
            print(f"[BD] Nie udao si wczyta danych dla {symbol}: {e}")

    return all_data
```

```python
def create_named_dataframes(stock_data, symbols): 
    """
    Tworzy osobne DataFrame'y dla kadego symbolu w postaci sownika
    i jednoczenie tworzy zmienne globalne df_SYMBOL, np. df_AAPL.
    Zwraca dict: {"AAPL": df_AAPL, ...}
    """
    named_dfs = {}

    for symbol in symbols:
        if symbol in stock_data:
            df = stock_data[symbol].copy()
            named_dfs[symbol] = df

            # Tworzenie zmiennej globalnej np. df_AAPL
            var_name = f"df_{symbol.replace('.', '_')}"
            globals()[var_name] = df
        else:
            print(f"[UWAGA] Brak danych dla: {symbol}")

    return named_dfs
```

## Funkcja obliczania wskanikw technicznych

```python
def prepare_nhits_lstm_tft_data(
    df,
    unique_id_value,
    price_col='Close',
    sma_range=range(2, 151),
    dropna=True
):
    """
    Przygotowuje DataFrame pod modele NHITS / LSTM / TFT z rolling feature'ami:
    - SMA, Slope
    - Price Momentum, Price Acceleration
    - RSI
    - VIX (annualizowana rolling volatilno)
    - Return Rate (procentowa stopa zwrotu)
    - Percent Off 52-Week High (zawsze z 252 dni)
    Wszystkie feature'y rolling dla okien w sma_range.
    Zwraca: (df_model, futr_exog_list)
    """

    # Daty (indeks lub kolumna)
    df = df.copy()
    if 'ds' in df.columns:
        date_values = pd.to_datetime(df['ds'])
        df = df.set_index('ds')
    else:
        date_values = df.index
        df.index = pd.to_datetime(df.index)

    exog_dict = {}
    futr_exog_list = []

    # Rolling log-returns (do VIX)
    log_returns = np.log(df[price_col] / df[price_col].shift(1))

    for window in sma_range:
        # SMA
        sma_col = f'MA_{window}'
        exog_dict[sma_col] = df[price_col].rolling(window).mean()
        futr_exog_list.append(sma_col)

        # Slope
        slope_col = f'MA_{window}_slope'
        exog_dict[slope_col] = exog_dict[sma_col].diff()
        futr_exog_list.append(slope_col)

        # Momentum
        momentum_col = f'price_momentum_{window}'
        exog_dict[momentum_col] = df[price_col].diff(window)
        futr_exog_list.append(momentum_col)

        # Acceleration
        acceleration_col = f'price_acceleration_{window}'
        exog_dict[acceleration_col] = df[price_col].diff(window).diff()
        futr_exog_list.append(acceleration_col)

        # RSI
        delta = df[price_col].diff()
        up = delta.clip(lower=0)
        down = -delta.clip(upper=0)
        roll_up = up.rolling(window).mean()
        roll_down = down.rolling(window).mean()
        rs = roll_up / roll_down
        rsi_col = f'RSI_{window}'
        exog_dict[rsi_col] = 100 - (100 / (1 + rs))
        futr_exog_list.append(rsi_col)

        # VIX (annualized rolling volatility z log zwrotw)
        vix_col = f'VIX_{window}'
        exog_dict[vix_col] = log_returns.rolling(window).std() * np.sqrt(252)
        futr_exog_list.append(vix_col)

        # Return Rate
        return_rate_col = f'return_rate_{window}'
        exog_dict[return_rate_col] = (df[price_col] / df[price_col].shift(window) - 1) * 100
        futr_exog_list.append(return_rate_col)

    # Percent Off 52-Week High (rolling)
    high_52w_window = 252
    high_52w_col = f'percent_off_52w_high'
    high_52w = df[price_col].rolling(window=high_52w_window, min_periods=1).max()
    exog_dict[high_52w_col] = 100 * (df[price_col] - high_52w) / high_52w
    futr_exog_list.append(high_52w_col)

    # Budujemy finalny DataFrame
    df_exog = pd.DataFrame(exog_dict, index=df.index)
    df_model = pd.DataFrame({
        'unique_id': unique_id_value,
        'ds': date_values,
        'y': df[price_col],
    }, index=df.index)
    df_model = pd.concat([df_model, df_exog], axis=1)

    if dropna:
        df_model = df_model.dropna(how='any').reset_index(drop=True)
    else:
        df_model = df_model.reset_index(drop=True)

    # Step: kolejne liczby od 0 (dla atwiejszej referencji)
    df_model['step'] = range(len(df_model))

    # Upewnij si, e 'ds' to datetime
    if not pd.api.types.is_datetime64_any_dtype(df_model['ds']):
        df_model['ds'] = pd.to_datetime(df_model['ds'])

    return df_model, futr_exog_list
```

## Funkcja Modelu

```python
def prediction_model(
    df_full,
    futr_exog_list,
    h=10,
    input_size=150,
    train_ratio=0.7,
    val_ratio=0.15,
    model_save_path="C:/Users/pawel/Documents/Python nauka/Projekt/Modele_predykcji_akcji",
    max_steps=2500,
    folds=3,
    rolling_window=0.1,
    show_learning_curve=True,
    dry_run=False,
    monitor_time=True,
    verbose=True,
    permutation_importance=True    # <-- DODANE, by mc sterowa
):
    time0 = time.time()

    if not os.path.exists(model_save_path):
        os.makedirs(model_save_path, exist_ok=True)

    n = len(df_full)
    train_end = int(n * train_ratio)
    val_end = train_end + int(n * val_ratio)
    df_train = df_full.iloc[:train_end].reset_index(drop=True)
    df_val = df_full.iloc[train_end:val_end].reset_index(drop=True)
    df_test = df_full.iloc[val_end:].reset_index(drop=True)

    if verbose:
        print(f"[Split] Train: {len(df_train)} | Val: {len(df_val)} | Test: {len(df_test)}")

    # ---- DRY RUN ----
    if dry_run:
        print("\n[Dry run] Pomijam trenowanie, tylko testuj pipeline i sprawdzam podzia danych.")
        return {
            "df_train": df_train,
            "df_val": df_val,
            "df_test": df_test,
            "info": "Dry run - bez trenowania modelu"
        }

    # ---- K-FOLD (rolling window) ----
    cv_scores = []
    cv_losses = []
    est_total_cv = None
    if folds > 1:
        if verbose:
            print(f"\n== [CV] Rolling window cross-validation: {folds} foldw ==")
        fold_size = int(len(df_train) * rolling_window)
        for fold in range(folds):
            fold_start = fold * fold_size // 2
            fold_train = df_train.iloc[:fold_start + fold_size]
            fold_val = df_train.iloc[fold_start + fold_size : fold_start + fold_size + fold_size]
            if len(fold_val) < 10:
                break
            time_fold_start = time.time()
            model_cv = NeuralForecast(
                models=[NHITS(
                    h=h,
                    input_size=input_size,
                    futr_exog_list=futr_exog_list,
                    max_steps=max_steps,
                    enable_checkpointing=False
                )],
                freq='B'
            )
            model_cv.fit(df=fold_train)
            expected_futr_val = model_cv.make_future_dataframe(df=fold_train)
            futr_exog_cv = pd.concat([fold_train, fold_val], ignore_index=True)[['ds', 'unique_id'] + futr_exog_list]
            futr_df_cv = expected_futr_val.merge(futr_exog_cv, on=['ds', 'unique_id'], how='left')
            futr_df_cv[futr_exog_list] = futr_df_cv[futr_exog_list].ffill().bfill()
            y_true_cv = fold_val.set_index('ds')['y']
            forecast_cv = model_cv.predict(futr_df=futr_df_cv)
            # --- KLUCZOWA POPRAWKA: tylko wsplne daty ---
            common_idx_cv = y_true_cv.index.intersection(forecast_cv['ds'])
            y_true_cv_valid = y_true_cv.loc[common_idx_cv]
            y_pred_cv_valid = forecast_cv.set_index('ds')['NHITS'].loc[common_idx_cv]
            mae_cv = mean_absolute_error(y_true_cv_valid, y_pred_cv_valid)
            rmse_cv = np.sqrt(mean_squared_error(y_true_cv_valid, y_pred_cv_valid))
            time_fold = time.time() - time_fold_start
            cv_scores.append({'fold': fold, 'mae': mae_cv, 'rmse': rmse_cv, 'time': time_fold})
            if hasattr(model_cv.models[0], "losses"):
                cv_losses.append(model_cv.models[0].losses)
            if verbose:
                print(f"[CV fold {fold+1}] MAE: {mae_cv:.4f}, RMSE: {rmse_cv:.4f}, Time: {cv_scores[-1]['time']:.1f}s")
            if fold == 0 and monitor_time:
                est_total_cv = time_fold * folds
                print(f"[Szacowany czas CV] Wszystkie foldy: ~{est_total_cv/60:.1f} min ({est_total_cv:.1f} s)")
        print(f"[CV] redni MAE: {np.mean([s['mae'] for s in cv_scores]):.4f}, redni RMSE: {np.mean([s['rmse'] for s in cv_scores]):.4f}")
        # --- Krzywe uczenia dla foldw ---
        if show_learning_curve and len(cv_losses) > 0:
            plt.figure(figsize=(12, 5))
            for i, losses in enumerate(cv_losses):
                plt.plot(losses, label=f'Fold {i+1}')
            plt.title("Krzywe uczenia NHITS (train loss, cross-validation)")
            plt.xlabel("Iteracja")
            plt.ylabel("Strata")
            plt.legend()
            plt.show()

    # ---- GWNY MODEL: train/val/test ----
    model_file = os.path.join(model_save_path, "neuralforecast_model.pkl")
    model_exists = os.path.exists(model_file)

    t1 = time.time()
    if verbose: print("\n== [1] Trening (train), walidacja (val) ==")
    if model_exists:
        if verbose: print(f"Model ju istnieje: {model_file}  wczytuj...")
        model_val = NeuralForecast.load(path=model_save_path)
    else:
        if verbose: print(f"Model nie istnieje  trenuj i zapisuj do: {model_file}")
        model_val = NeuralForecast(
            models=[NHITS(
                h=h,
                input_size=input_size,
                futr_exog_list=futr_exog_list,
                max_steps=max_steps,
                enable_checkpointing=False
            )],
            freq='B'
        )
        train_start = time.time()
        model_val.fit(df=df_train)
        train_time = time.time() - train_start
        if monitor_time:
            total_steps = 3
            est_total_train = train_time * total_steps
            print(f"[Szacowany czas gwnych etapw] (bez CV): ~{est_total_train/60:.1f} min ({est_total_train:.1f} s)")
        model_val.save(path=model_save_path, overwrite=True, save_dataset=True)
    t2 = time.time()
    print(f"[Czas treningu (train/val)]: {t2-t1:.1f} s")

    # --- Krzywa uczenia gwnego modelu ---
    if show_learning_curve and hasattr(model_val.models[0], "losses"):
        try:
            losses = model_val.models[0].losses
            plt.figure(figsize=(10, 4))
            plt.plot(losses, label='train loss (val model)')
            plt.title("NHITS - krzywa uczenia (train loss, walidacja)")
            plt.xlabel("Iteracja")
            plt.ylabel("Strata")
            plt.legend()
            plt.show()
        except Exception as e:
            print(f"Nie udao si wywietli krzywej uczenia: {e}")

    # Egzogeniczne na walidacj
    df_futr_fill = pd.concat([df_train, df_val], ignore_index=True)[['ds', 'unique_id'] + futr_exog_list]
    expected_futr_val = model_val.make_future_dataframe(df=df_train)
    futr_df_val = expected_futr_val.merge(df_futr_fill, on=['ds', 'unique_id'], how='left')
    futr_df_val[futr_exog_list] = futr_df_val[futr_exog_list].ffill().bfill()
    forecast_val = model_val.predict(futr_df=futr_df_val)
    y_true_val = df_val.set_index('ds')['y']
    common_idx_val = y_true_val.index.intersection(forecast_val['ds'])
    y_true_val_valid = y_true_val.loc[common_idx_val]
    y_pred_val_valid = forecast_val.set_index('ds')['NHITS'].loc[common_idx_val]
    mae_val = mean_absolute_error(y_true_val_valid, y_pred_val_valid)
    rmse_val = np.sqrt(mean_squared_error(y_true_val_valid, y_pred_val_valid))
    if verbose:
        print(f"[VAL] MAE: {mae_val:.4f}, RMSE: {rmse_val:.4f}, dni: {len(y_true_val_valid)}")

    # === Permutation Feature Importance ===
    if permutation_importance:
        print("\n=== Permutation Feature Importance na walidacji ===")
        baseline_pred = forecast_val.set_index('ds')['NHITS']
        y_true = df_val.set_index('ds')['y']
        common_idx = y_true.index.intersection(baseline_pred.index)
        baseline_rmse = np.sqrt(mean_squared_error(y_true.loc[common_idx], baseline_pred.loc[common_idx]))
        print(f"RMSE baseline: {baseline_rmse:.4f}")
        feature_groups = {
            'SMA': [c for c in futr_exog_list if c.startswith('MA_') and not c.endswith('_slope')],
            'Slope': [c for c in futr_exog_list if c.endswith('_slope')],
            'Momentum': [c for c in futr_exog_list if 'momentum' in c],
            'Acceleration': [c for c in futr_exog_list if 'acceleration' in c],
            'RSI': [c for c in futr_exog_list if c.startswith('RSI')],
            'VIX': [c for c in futr_exog_list if c.startswith('VIX')],
            'Return Rate': [c for c in futr_exog_list if 'return_rate' in c],
            'Percent Off 52-Week High': [c for c in futr_exog_list if 'percent_off_52w_high' in c]
        }
        results = {}
        for group, columns in feature_groups.items():
            futr_df_val_permuted = futr_df_val.copy()
            for col in columns:
                futr_df_val_permuted[col] = np.random.permutation(futr_df_val_permuted[col].values)
            perm_pred = model_val.predict(futr_df=futr_df_val_permuted).set_index('ds')['NHITS']
            perm_rmse = np.sqrt(mean_squared_error(y_true.loc[common_idx], perm_pred.loc[common_idx]))
            delta = perm_rmse - baseline_rmse
            results[group] = delta
            print(f"{group:<30} RMSE: {delta:.4f}")
        print("\n--- Wpyw na RMSE po permutacji feature'w (wiksze  = waniejsze) ---")
        for k, v in sorted(results.items(), key=lambda x: -x[1]):
            print(f"{k:<30} RMSE: {v:.4f}")

    # ---- Retrain na train+val, test ----
    t3 = time.time()
    if verbose: print("\n== [2] Retrain (train+val), test ==")
    df_trainval = pd.concat([df_train, df_val], ignore_index=True)
    retrain_start = time.time()
    model_test = NeuralForecast(
        models=[NHITS(
            h=h,
            input_size=input_size,
            futr_exog_list=futr_exog_list,
            max_steps=max_steps,
            enable_checkpointing=False
        )],
        freq='B'
    )
    model_test.fit(df=df_trainval)
    retrain_time = time.time() - retrain_start
    print(f"[Czas retrain (train+val)]: {retrain_time:.1f} s")
    model_test.save(path=model_save_path, overwrite=True, save_dataset=True)
    t4 = time.time()

    if show_learning_curve and hasattr(model_test.models[0], "losses"):
        try:
            losses = model_test.models[0].losses
            plt.figure(figsize=(10, 4))
            plt.plot(losses, label='train loss (retrain)')
            plt.title("NHITS - krzywa uczenia (train loss, retrain)")
            plt.xlabel("Iteracja")
            plt.ylabel("Strata")
            plt.legend()
            plt.show()
        except Exception as e:
            print(f"Nie udao si wywietli krzywej uczenia (retrain): {e}")

    expected_futr_test = model_test.make_future_dataframe(df=df_trainval)
    futr_df_test = expected_futr_test.merge(
        df_test[['ds', 'unique_id'] + futr_exog_list],
        on=['ds', 'unique_id'],
        how='left'
    )
    futr_df_test[futr_exog_list] = futr_df_test[futr_exog_list].ffill().bfill()
    forecast_test = model_test.predict(futr_df=futr_df_test)
    y_pred_test = forecast_test.set_index('ds')['NHITS']
    y_true_test = df_test.set_index('ds')['y']
    common_idx_test = y_pred_test.index.intersection(y_true_test.index)
    y_true_test = y_true_test.loc[common_idx_test]
    y_pred_test = y_pred_test.loc[common_idx_test]
    mae_test = mean_absolute_error(y_true_test, y_pred_test)
    rmse_test = np.sqrt(mean_squared_error(y_true_test, y_pred_test))
    if verbose:
        print(f"[TEST] MAE: {mae_test:.4f}, RMSE: {rmse_test:.4f}, dni: {len(y_true_test)}")

    # ---- Predykcja na przyszo ----
    t5 = time.time()
    if verbose: print("\n== [3] Predykcja na przyszo ==")
    future_start = time.time()
    df_full = df_full.reset_index(drop=True)
    model_future = NeuralForecast(
        models=[NHITS(
            h=h,
            input_size=input_size,
            futr_exog_list=futr_exog_list,
            max_steps=max_steps,
            enable_checkpointing=False
        )],
        freq='B'
    )
    model_future.fit(df=df_full)
    future_time = time.time() - future_start
    print(f"[Czas prognozy przyszo]: {future_time:.1f} s")
    model_future.save(path=model_save_path, overwrite=True, save_dataset=True)
    t6 = time.time()

    if show_learning_curve and hasattr(model_future.models[0], "losses"):
        try:
            losses = model_future.models[0].losses
            plt.figure(figsize=(10, 4))
            plt.plot(losses, label='train loss (future)')
            plt.title("NHITS - krzywa uczenia (train loss, future)")
            plt.xlabel("Iteracja")
            plt.ylabel("Strata")
            plt.legend()
            plt.show()
        except Exception as e:
            print(f"Nie udao si wywietli krzywej uczenia (future): {e}")

    expected_future = model_future.make_future_dataframe(df=df_full)
    last_exog_row = df_full[futr_exog_list].iloc[-1]
    exog_future = pd.DataFrame([last_exog_row.values] * len(expected_future), columns=futr_exog_list).reset_index(drop=True)
    futr_df_future = pd.concat([expected_future.reset_index(drop=True), exog_future], axis=1)
    futr_df_future['unique_id'] = futr_df_future['unique_id'].astype(str)
    futr_df_future['ds'] = pd.to_datetime(futr_df_future['ds'])
    futr_df_future = futr_df_future[['unique_id', 'ds'] + futr_exog_list].reset_index(drop=True)
    forecast_future = model_future.predict(df=df_full, futr_df=futr_df_future)

    # Automatyczne sprawdzanie przeuczenia
    overfitting_detected = False
    if mae_val < 0.9 * mae_test:
        overfitting_detected = True
        if verbose:
            print("\n[OSTRZEENIE] Moliwe przeuczenie: MAE walidacyjny duo niszy ni testowy!")

    # Wizualizacja
    if verbose:
        plt.figure(figsize=(14, 6))
        plt.plot(y_true_test.index, y_true_test.values, label='Rzeczywiste', marker='o')
        plt.plot(y_pred_test.index, y_pred_test.values, label='Prognoza NHITS (test)', marker='x')
        plt.title('Test: Prognoza vs rzeczywiste')
        plt.legend()
        plt.grid(True)
        plt.show()

        plt.figure(figsize=(14, 6))
        plt.plot(df_full['ds'], df_full['y'], label='Historyczne dane', alpha=0.6)
        plt.plot(forecast_future['ds'], forecast_future['NHITS'], label='Prognoza na przyszo', marker='x')
        plt.title('Prognoza NHITS  okres przyszy')
        plt.legend()
        plt.grid(True)
        plt.show()

        # Ostatnie 10 dni + 10 dni prognozy
        history_tail = df_full.tail(10).copy()
        forecast_future_display = forecast_future.copy()
        df_display = pd.merge(
            history_tail[['ds', 'y']],
            forecast_future_display[['ds', 'NHITS']],
            on='ds',
            how='outer'
        ).sort_values('ds').reset_index(drop=True)

        plt.figure(figsize=(14, 6))
        plt.plot(history_tail['ds'], history_tail['y'], label='Rzeczywiste (ostatnie 10)', marker='o')
        plt.plot(forecast_future_display['ds'], forecast_future_display['NHITS'], label='Prognoza NHITS (kolejne 10)', marker='x')
        plt.title('Ostatnie 10 dni + Prognoza na przyszo (10 dni)')
        plt.xlabel('Data')
        plt.ylabel('y')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        from IPython.display import display
        print("\nTabela: ostatnie 10 dni + prognoza 10 dni:")
        display(df_display.round(2))

    # Wywietlenie czasw
    if monitor_time:
        print(f"\nCzasy etapw:")
        print(f"Trening (train): {t2-t1:.1f} s | Retrain (train+val): {t4-t3:.1f} s | Prognoza przyszo: {t6-t5:.1f} s | Razem: {time.time()-time0:.1f} s")
        if est_total_cv:
            print(f"[Szacowany czas cross-validation (CV)]: ~{est_total_cv/60:.1f} min ({est_total_cv:.1f} s)")
        if monitor_time and 'est_total_train' in locals():
            print(f"[Szacowany czas gwnych etapw (bez CV)]: ~{est_total_train/60:.1f} min ({est_total_train:.1f} s)")

    return {
        "model_val": model_val,
        "forecast_val": forecast_val,
        "mae_val": mae_val,
        "rmse_val": rmse_val,
        "model_test": model_test,
        "forecast_test": forecast_test,
        "mae_test": mae_test,
        "rmse_test": rmse_test,
        "model_future": model_future,
        "forecast_future": forecast_future,
        "df_train": df_train,
        "df_val": df_val,
        "df_test": df_test,
        "overfitting_detected": overfitting_detected,
        "cv_scores": cv_scores,
    }
```
